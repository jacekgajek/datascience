```python
import plotly.express as px
import numpy as np


def f(x):
    return  0.005 * x**4 - 2*x**3 + 7 *  x - 5

def f_1(x):
    return 0.020 * x**3 - 6* x**2 + 7

def tangent(x, x0):
    der = f_1(x0)
    return der * x + f(x0) - x0*der


X = np.linspace(-4, 4, 200)
Y = [ f(x) for x in X ]

tang = [ tangent(x, -2) for x in X ]
# print(matrixNorm(M))


fig = px.line(x=X, y=[Y, tang])

fig.show()
```
