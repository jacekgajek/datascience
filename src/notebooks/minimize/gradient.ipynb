```python
import numpy as np
import plotly.express as px

# def f(x):
#     return -x*(x-1)*(x-2)*(x+1)*(x+2)
def f(x):
    return -x**5 + 2*x**3 - 3*x**2 + 2*x

def df(x) :
    return -5*x**4 + 6*x**2 - 6*x + 2

def minimize(f, df, start, end):
    h = 0.01
    min_delta = 0.001
    vx = np.random.uniform(start, end)
    print(vx)
    i = 0

    while True:
        # print(f"v = {vx}, f(vx) = {f(vx)} dv = {df(vx)}")
        i += 1
        delta = - h*df(vx) #np.sign(df(vx))
        vx = vx + delta
        if vx >= end or vx <= start or abs(delta) <= min_delta:
            break
    print(f"x = {vx}, f(x) = {f(vx)} dv = {df(vx)}, steps: {i}")
    return vx

start = -2
end = 1.5
lowerLimit = -10
upperLimit = 10

X = np.linspace(start, end, 100)
Y1 = [f(x) for x in X]

minimize(f, df, start, end)

fig = px.line(x=X, y=Y1, range_x =[start, end], range_y=[lowerLimit,upperLimit])

fig.show()
```
